# Calculadora de probabilidades - T DE STUDENT




```python
# Cargamos librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Distribución t de Student

$t=\frac{Z}{\sqrt{\frac{U}{n}}}\sim t_n,\quad \text{donde} \qquad Z \sim N(0,1) \quad y \quad U\sim \chi_n^2$

**Cambiar la siguiente celda para calcular probabilidades**

```python
# Parámetros
g = 10
# Valores de t
t1 = 1.5  # Valor para P(T < t1)
t2 = 2.0  # Valor para P(T > t2)
t3_intervalo = -1.0  # Límite inferior para P(t3 < T < t4)
t4_intervalo = 1.5   # Límite superior para P(t3 < T < t4)

```

**Cambiar la celda de arriba para calcular probabilidades**

```python
# Probabilidad en la cola izquierda: P(T < t1)
prob_izq_t1 = stats.t.cdf(t1, df=g) #t.cdf es la función de distribución
prob_izq_t1
```

```python
# Graficamos sombreando la probabilidad buscada
t_values = np.linspace(-4, 4, 400)
pdf_values = stats.t.pdf(t_values, df=g)

# Grafico
plt.figure(figsize=(8, 4))
plt.plot(t_values, pdf_values, color='blue') # Graficar la distribución t de Student
# Sombrear la región P(T < t1)
t1_values = np.linspace(-4, t1, 200)
plt.fill_between(t1_values, 0, stats.t.pdf(t1_values, df=g), color='green', alpha=0.3)
plt.title("Distribución t de Student")
plt.xlabel("Valores de t")
plt.ylabel("Densidad")
```

```python
# Probabilidad en la cola derecha: P(T > t2)
prob_der_t2 = 1 - stats.t.cdf(t2, df=g)

prob_der_t2
```

```python
# Graficamos sombreando la probabilidad buscada
t_values = np.linspace(-4, 4, 400)
pdf_values = stats.t.pdf(t_values, df=g)

# Grafico
plt.figure(figsize=(8, 4))
plt.plot(t_values, pdf_values, color='blue')
# Sombrear la región P(T > t2)
t2_values = np.linspace(t2, 4, 200)
plt.fill_between(t2_values, 0, stats.t.pdf(t2_values, df=g), color='orange', alpha=0.3)
plt.title("Distribución t de Student")
plt.xlabel("Valores de t")
plt.ylabel("Densidad")
```

```python
# Probabilidad en un intervalo P(t3 < T < t4)
prob_entre_t3_t4 = stats.t.cdf(t4_intervalo, df=g) - stats.t.cdf(t3_intervalo, df=g)
prob_entre_t3_t4
```

```python
# Graficamos sombreando la probabilidad buscada
t_values = np.linspace(-4, 4, 400)
pdf_values = stats.t.pdf(t_values, df=g)

# Grafico
plt.figure(figsize=(8, 4))
plt.plot(t_values, pdf_values, color='blue') # Graficar la distribución t de Student
# Sombreamos el área P(t1_interval < T < t2_interval)
intervalo_values = np.linspace(t3_intervalo, t4_intervalo, 200)
plt.fill_between(intervalo_values, 0, stats.t.pdf(intervalo_values, df=g), color='purple', alpha=0.3)
plt.title("Distribución t de Student")
plt.xlabel("Valores de t")
plt.ylabel("Densidad")
```

**Cambiar la siguiente celda para calcular cuantiles** (valores de la variable)





```python
# Parámetros
g = 14 #grados de libertad
# Probabilidades
p1 = 0.975  # Probabilidad para P(T < t1)
p2 = 0.1  # Probabilidad para P(T > t2)
p_intervalo = 0.8  # Probabilidad para P(t1 < T < t2)
```

**Cambiar la celda de arriba para calcular cuantiles** (valores de la variable)


```python
# Calcular t1 tal que P(T < t1) = p1
t1 = stats.t.ppf(p1, df=g) #donde ppf es la función de cuantiles inversa
t1
```

```python
# Calcular t2 tal que P(T > t2) = p2
t2 = stats.t.ppf(1 - p2, df=g)
t2
```

```python
# Calcular t3 y t4 tal que P(t3 < T < t4) = p_intervalo
t3_intervalo = stats.t.ppf((1 - p_intervalo) / 2, df=g) # por simetria se tiene (1 - p_intervalo) / 2 en cada cola
t4_intervalo = stats.t.ppf(1 - (1 - p_intervalo) / 2, df=g)

print(f"t3 y t4 tal que P({t3_intervalo:.4f} < T < {t4_intervalo:.4f}) = {p_intervalo} son t3 = {t3_intervalo:.4f} y t4 = {t4_intervalo:.4f}")
```
