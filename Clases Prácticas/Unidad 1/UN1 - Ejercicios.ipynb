## Distribución Chi cuadrado (X2)









Se define como la suma de n variables normales estándar.

$W=\sum_{i=1}^n Z_i^2 = \sum_{i=1}^n \frac{(X_i-\mu)^2}{\sigma^2} ∼ \chi_n^2$

Si no se conoce la media poblacional: $\sum_{i=1}^n \frac{(X_i-\bar{X})^2}{n-1}\sim \chi_{n-1}^2$

Por su parte, el estadístico $\frac{(n-1)S^2}{\sigma^2} \sim \chi_{n-1}^2$ es aquel que relaciona la varianza poblacional y la muestral.

## Distribución t de Student

$t=\frac{Z}{\sqrt{\frac{U}{n}}}\sim t_n,\quad \text{donde} \qquad Z \sim N(0,1) \quad y \quad U\sim \chi_n^2$

###  Ejercicio. Distribución t de Student

El estadístico k con distribución t de Student con 20 grados de libertad se utiliza para medir la capacidad que tiene el precio de un bien sustituto para explicar los cambios en la demanda.
Si k toma valores dentro del intervalo:  -1,725 < t <  1,725 se confirma que el precio tiene capacidad para explicar los cambios en la demanda. Se pide:


1.   ¿Cuál es la probabilidad que el precio no tenga dicha capacidad?

2.   ¿Cuál debería ser el intervalo de k (valor inferior y superior), para que la probabilidad de explicar los cambios en la demanda sea 0,99?




```python
# Parámetros
g = 20  # Grados de libertad
t1 = -1.725
t2 = 1.725
p_intervalo = 0.99  # Probabilidad deseada
```

```python
# 1. Probabilidad que el precio no tenga capacidad para explicar los cambios en la demanda, probabilidad fuera del intervalo P(t1 < T < t2)
prob_no_explicar = stats.t.cdf(t1, df=g) + (1 - stats.t.cdf(t2, df=g)) # P(T < t1) + P(T > t2)
prob_no_explicar
```

```python
# 2. Intervalo de k para que la probabilidad de explicar los cambios en la demanda sea 0.99, [t1,t2] tal que P(t1 < T < t2) = 0.99

t1_intervalo = stats.t.ppf((1 - p_intervalo) / 2, df=g)
t2_intervalo = stats.t.ppf(1 - (1 - p_intervalo) / 2, df=g)
print(f"t1 = {t1_intervalo:.4f} y t2 = {t2_intervalo:.4f}")
```

## Distribución F de Snedecor

Es el cociente de dos variables chi cuadradao estadísticamente independientes, divididas por sus grados de libertad, resp.

$$\begin{align*}
U&\sim \chi^2_n, & V&\sim\chi^2_n,
\end{align*}$$

$F = \frac{\frac{U}{n}}{\frac{V}{m}} $
