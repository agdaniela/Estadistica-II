# Calculadora de probabilidades - CHI CUADRADO


```python
# Cargamos librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Distribución Chi cuadrado ($\chi^2$)









Se define como la suma de n variables normales estándar.

$W=\sum_{i=1}^n Z_i^2 = \sum_{i=1}^n \frac{(X_i-\mu)^2}{\sigma^2} ∼ \chi_n^2$

Si no se conoce la media poblacional: $\sum_{i=1}^n \frac{(X_i-\bar{X})^2}{n-1}\sim \chi_{n-1}^2$

Por su parte, el estadístico $\frac{(n-1)S^2}{\sigma^2} \sim \chi_{n-1}^2$ es aquel que relaciona la varianza poblacional y la muestral.

**Cambiar la siguiente celda para calcular probabilidades!!!**

```python
# Parámetros
g = 10 #grados de libertad
# Valores de X
x1 = 2  # Valor para P(X < x1)
x2 = 6  # Valor para P(X > x2)
x3_intervalo = 7.0  # Límite inferior para P(x3 < X < x4)
x4_intervalo = 10.0   # Límite superior para P(x3 < X < x4)

```

**Cambiar la celda de arriba para calcular probabilidades!**

```python
# Probabilidad en la cola izquierda: P(X < x1)
prob_izq_x1 = stats.chi2.cdf(x1, df=g) #chi2.cdf es la función de distribución
prob_izq_x1
```

```python
# Graficamos sombreando la probabilidad buscada
chi_values = np.linspace(0, 100, 400)
pdf_values = stats.chi2.pdf(chi_values, df=g)


# Grafico
plt.figure(figsize=(8, 4))
plt.plot(chi_values, pdf_values, color='blue') # Graficar la distribución chi cuadrado
# Sombrear la región P(X < x1)
x1_values = np.linspace(0, x1, 200)
plt.fill_between(x1_values, stats.chi2.pdf(x1_values, g), alpha=0.3, color='green', label=f'Área acumulada a la izquierda de x={x1}')
plt.title("Distribución Chi Cuadrado")
plt.xlabel("Valores de X")
plt.ylabel("Densidad")
```

```python
# Probabilidad en la cola derecha: P(X > x2)
prob_der_x2 = 1 - stats.chi2.cdf(x2, df=g)

prob_der_x2
```

```python
# Graficamos sombreando la probabilidad buscada
chi_values = np.linspace(0, 100, 400)
pdf_values = stats.chi2.pdf(chi_values, df=g)


# Grafico
plt.figure(figsize=(8, 4))
plt.plot(chi_values, pdf_values, color='blue') # Graficar la distribución chi cuadrado
# Sombrear la región P(X > x2)
x2_values = np.linspace(x2, 10, 200)
plt.fill_between(x2_values, 0, stats.chi2.pdf(x2_values, g), alpha=0.3, color='orange', label=f'Área acumulada a la derecha de x={x2}')
plt.title("Distribución Chi Cuadrado")
plt.xlabel("Valores de X")
plt.ylabel("Densidad")
```

```python
# Probabilidad en un intervalo P(x3 < X < x4)
prob_entre_x3_x4 = stats.chi2.cdf(x4_intervalo, df=g) - stats.chi2.cdf(x3_intervalo, df=g)
prob_entre_x3_x4
```

```python
# Graficamos sombreando la probabilidad buscada
chi_values = np.linspace(0, 100, 400)
pdf_values = stats.chi2.pdf(chi_values, df=g)


# Grafico
plt.figure(figsize=(8, 4))
plt.plot(chi_values, pdf_values, color='blue') # Graficar la distribución chi cuadrado
# Sombrear la región P(X > x1)
x_intervalo = np.linspace(x3_intervalo, x4_intervalo, 200)
plt.fill_between(x_intervalo, stats.chi2.pdf(x_intervalo, g), alpha=0.3, color='purple')
plt.title("Distribución Chi Cuadrado")
plt.xlabel("Valores de X")
plt.ylabel("Densidad")
```

**Cambiar la siguiente celda para calcular cuantiles** (valores de la variable)




```python
# Parámetros
g = 10  # grados de libertad
# Probabilidades
p1 = 0.9 # Probabilidad para P(X < x1)
p2 = 0.1  # Probabilidad para P(X > x2)
p_intervalo = 0.8  # Probabilidad para P(x1 < X < x2)
```

**Cambiar la celda de arriba para calcular cuantiles** (valores de la variable)





```python
# Calcular x1 tal que P(X < x1) = p1
x1 = stats.chi2.ppf(p1, df=g) #donde ppf es la función de cuantiles inversa
x1
```

```python
# Calcular x2 tal que P(X > x2) = p2
x2 = stats.chi2.ppf(1 - p2, df=g)
x2
```
