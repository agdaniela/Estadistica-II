# Calculadora de probabilidades - NORMAL ESTÁNDAR




```python
# Cargamos librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Distribución Normal

Distribución normal
$$\begin{align*}
X&\sim N(\mu,\sigma)
\end{align*}$$

Distribución normal estándar
$$\begin{align*}
Z&\sim N(0,1)
\end{align*}$$

**Cambiar la siguiente celda para calcular probabilidades**

```python
# Valores de Z
z1 = -1  # Valor para P(Z < z1)
z2 = 1  # Valor para P(Z > z2)
z3_intervalo = -2.0  # Límite inferior para P(z3 < Z < z4)
z4_intervalo = 2.0   # Límite superior para P(z3 < Z < z4)

```

**Cambiar la celda de arriba para calcular probabilidades**

```python
# Probabilidad en la cola izquierda: P(Z < z1)
prob_izq_z1 = stats.norm.cdf(z1)
prob_izq_z1
```

```python
# Graficamos sombreando la probabilidad buscada
z_values = np.linspace(-4, 4, 400)
pdf_values = stats.norm.pdf(z_values)

# Grafico
plt.figure(figsize=(8, 4))
plt.plot(z_values, pdf_values, color='blue') # Graficar la distribución
# Sombrear la región P(Z < z1)
z1_values = np.linspace(-4, z1, 200)
plt.fill_between(z1_values, 0, stats.norm.pdf(z1_values), color='green', alpha=0.3)
plt.title("Distribución Normal")
plt.xlabel("Valores de z")
plt.ylabel("Densidad")
```

```python
# Probabilidad en la cola derecha: P(Z > z2)
prob_der_z2 = 1 - stats.norm.cdf(z2)

prob_der_z2
```

```python
# Graficamos sombreando la probabilidad buscada
z_values = np.linspace(-4, 4, 400)
pdf_values = stats.norm.pdf(z_values)

# Grafico
plt.figure(figsize=(8, 4))
plt.plot(z_values, pdf_values, color='blue')
# Sombrear la región P(Z > z2)
z2_values = np.linspace(z2, 4, 200)
plt.fill_between(z2_values, 0, stats.norm.pdf(z2_values), color='orange', alpha=0.3)
plt.title("Distribución Normal")
plt.xlabel("Valores de z")
plt.ylabel("Densidad")
```

```python
# Probabilidad en un intervalo P(z3 < Z < z4)
prob_entre_z3_z4 = stats.norm.cdf(z4_intervalo) - stats.norm.cdf(z3_intervalo)
prob_entre_z3_z4
```

```python
# Graficamos sombreando la probabilidad buscada
z_values = np.linspace(-4, 4, 400)
pdf_values = stats.norm.pdf(z_values)

# Grafico
plt.figure(figsize=(8, 4))
plt.plot(z_values, pdf_values, color='blue') # Graficar la distribución
# Sombreamos el área P(z1_interval < Z < z2_interval)
intervalo_values = np.linspace(z3_intervalo, z4_intervalo, 200)
plt.fill_between(intervalo_values, 0, stats.norm.pdf(intervalo_values), color='purple', alpha=0.3)
plt.title("Distribución Normal")
plt.xlabel("Valores de z")
plt.ylabel("Densidad")
```

**Cambiar la siguiente celda para calcular cuantiles** (valores de la variable)





```python
# Probabilidades
p1 = 0.95  # Probabilidad para P(Z < z1)
p2 = 0.05  # Probabilidad para P(Z > z2)
p_intervalo = 0.95  # Probabilidad para P(z1 < Z < z2)
```

**Cambiar la celda de arriba para calcular cuantiles** (valores de la variable)


```python
# Calcular z1 tal que P(Z < z1) = p1
z1 = stats.norm.ppf(p1) #donde ppf es la función de cuantiles inversa
z1
```

```python
# Calcular z2 tal que P(Z > z2) = p2
z2 = stats.norm.ppf(1 - p2)
z2
```

```python
# Calcular z3 y z4 tal que P(z3 < Z < z4) = p_intervalo
z3_intervalo = stats.norm.ppf((1 - p_intervalo) / 2 ) # por simetria se tiene (1 - p_intervalo) / 2 en cada cola
z4_intervalo = stats.norm.ppf(1 - (1 - p_intervalo)/2 )

print(f"z3 y z4 tal que P(z3 < X < z4) = {p_intervalo} son z3 = {z3_intervalo:.4f} y z4 = {z4_intervalo:.4f}")
```
