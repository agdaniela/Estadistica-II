# Calculadora de probabilidades - F DE SNEDECOR




```python
# Cargamos librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Distribución F de Snedecor

Es el cociente de dos variables chi cuadrado estadísticamente independientes, divididas por sus grados de libertad, respectivamente.

$$\begin{align*}
U&\sim \chi^2_n, & V&\sim\chi^2_n,
\end{align*}$$

$F = \frac{\frac{U}{n}}{\frac{V}{m}} $

**Cambiar la siguiente celda para calcular probabilidades**

```python
# Parámetros
m = 30
n = 20
# Valores de X
x1 = 0.5  # Valor para P(X < x1)
x2 = 1.5  # Valor para P(X > x2)
x3_intervalo = 1.0  # Límite inferior para P(x3 < X < x4)
x4_intervalo = 2.0   # Límite superior para P(x3 < X < x4)
```

**Cambiar la celda de arriba para calcular probabilidades**

```python
# Probabilidad en la cola izquierda: P(X < x1)
prob_izq_x1 = stats.f.cdf(x1, m,n) #f.cdf es la función de distribución
prob_izq_x1
```

```python
# Graficamos sombreando la probabilidad buscada
f_values = np.linspace(0,5, 400)
pdf_values = stats.f.pdf(f_values, dfn=m,dfd=n)

# Grafico
plt.figure(figsize=(8, 4))
plt.plot(f_values, pdf_values, color='blue') # Graficar la distribución
# Sombrear la región P(X < x1)
x1_values = np.linspace(0, x1, 200)
plt.fill_between(x1_values, stats.f.pdf(x1_values, m,n), alpha=0.3, color='green', label=f'Área acumulada a la izquierda de x={x1}')
plt.title("Distribución F")
plt.xlabel("Valores de X")
plt.ylabel("Densidad")
```

```python
# Probabilidad en la cola derecha: P(X > x2)
prob_der_x2 = 1 - stats.f.cdf(x2, m,n)

prob_der_x2
```

```python
# Graficamos sombreando la probabilidad buscada
f_values = np.linspace(0,5, 400)
pdf_values = stats.f.pdf(f_values, dfn=m,dfd=n)


# Grafico
plt.figure(figsize=(8, 4))
plt.plot(f_values, pdf_values, color='blue') # Graficar la distribución
# Sombrear la región P(X > x1)
x2_values = np.linspace(x2, 5, 200)
plt.fill_between(x2_values, 0, stats.f.pdf(x2_values, m,n), alpha=0.3, color='orange', label=f'Área acumulada a la derecha de x={x2}')
plt.title("Distribución F")
plt.xlabel("Valores de X")
plt.ylabel("Densidad")
```

```python
# Probabilidad en un intervalo P(x3 < X < x4)
prob_entre_x3_x4 = stats.f.cdf(x4_intervalo, m,n) - stats.f.cdf(x3_intervalo, m,n)
prob_entre_x3_x4
```

```python
# Graficamos sombreando la probabilidad buscada
f_values = np.linspace(0,5, 400)
pdf_values = stats.f.pdf(f_values, dfn=m,dfd=n)


# Grafico
plt.figure(figsize=(8, 4))
plt.plot(f_values, pdf_values, color='blue') # Graficar la distribución
# Sombrear la región P(X > x1)
x_intervalo = np.linspace(x3_intervalo, x4_intervalo, 200)
plt.fill_between(x_intervalo, stats.f.pdf(x_intervalo, m,n), alpha=0.3, color='purple')
plt.title("Distribución F")
plt.xlabel("Valores de X")
plt.ylabel("Densidad")
```

**Cambiar la siguiente celda para calcular cuantiles** (valores de la variable)


```python
# Parámetros
m = 30
n = 20
# Probabilidades
p1 = 0.9  # Probabilidad para P(X < x1)
p2 = 0.1  # Probabilidad para P(X > x2)
```

**Cambiar la celda de arriba para calcular cuantiles** (valores de la variable)


```python
# Calcular x1 tal que P(X < x1) = p1

x1 = stats.f.ppf(p1, m,n)
x1
```

```python
# Calcular x2 tal que P(X > x2) = p2
x2 = stats.f.ppf(1-p2,m,n)
x2
```
